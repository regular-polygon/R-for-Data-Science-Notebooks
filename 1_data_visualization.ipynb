```r
penguins
```

### Create a visualization illustrating the relationship between flapper length and body mass for penguins, segmented by species.
The mapping argument defines how data should be mapped to the aesthetics of the plot.

```r
options(repr.plot.width = 10, repr.plot.height = 8)  # config ggplot output size

ggplot(
    data = penguins,
    mapping = aes(x = flipper_length_mm, y = body_mass_g)
    ) +
    geom_point(mapping = aes(color = species, shape = species), size = 3, na.rm = TRUE) +
    geom_smooth(method = "lm", na.rm = TRUE) +
      labs(
        title = "Body mass and flipper length",
        subtitle = "Dimensions for Adelie, Chinstrap, and Gentoo Penguins",
        x = "Flipper length (mm)", y = "Body mass (g)",
        color = "Species", shape = "Species",
        caption = "Data source: palmerpenguins dataset"
      ) +
      scale_color_colorblind() +
      theme(text=element_text(size=20))
```

### Make a scatterplot with bill_depth_mm on the y-axis and bill_length_mm on the x-axis.

```r
ggplot(
    data = penguins,
    mapping = aes(x = bill_length_mm, y = bill_depth_mm)
    ) +
    geom_point(na.rm = TRUE) +
    theme(text=element_text(size=20))
```

There appears to be a weak negative correlation between bill length and bill depth.

### Explore the relationship between species and bill_depth_mm

```r
ggplot(
    data = penguins,
    mapping = aes(x = species, y = bill_depth_mm)
    ) + 
    geom_boxplot(mapping = aes(color = species), na.rm = TRUE) +
    theme(text=element_text(size=20))
```

### Explore the relationship between flipper length and body mass with color scale based on bill depth

```r
ggplot(
    data = penguins,
    mapping = aes(x = flipper_length_mm, y = body_mass_g)
    ) + 
    geom_point(mapping = aes(color = bill_depth_mm), size = 4, na.rm = TRUE) +
    geom_smooth(se = TRUE, na.rm = TRUE) +
    theme(text=element_text(size=20))
```
